```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,LabelEncoder
```

```python
df = pd.read_csv("Dataset\Crop_recommendation.csv")
df.head(10)
```

```python
encoder = LabelEncoder()
df["label"]= encoder.fit_transform(df["label"])
df.head(5)
```

```python
input_dim = len(df.columns)-1
num_columns = len(df['label'].unique())
```

## Logistic Regression

```python
from logistic_regression import LogisticRegression

model = LogisticRegression(input_dim=input_dim,num_class=num_columns)
model.load_data(df)
model.train(epochs=200,batch_size=32,lr=0.01)
```
